```python
# Dividing the dataset into different variables.
# X is independent variable.
X = data.iloc[:, :4]
X
```

```python
# Y is dependent variable.
Y = data.iloc[:, 4]
Y
```

```python
# Converting categorical data points into numerical values.
# Performing one hot encoding using dummy variable trap.
states = pd.get_dummies(data['State'], drop_first=True)
states
```

```python
# Concatenating the data with dummy variables.
X = pd.concat([X, states],axis = 1).drop('State',axis = 1)
X
```

```python
# Splitting the dataset into training and testing set to perform linear regression.
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
# Returns values of rows and columns present in X.
X.shape
```

```python
# Returns values of rows and columns present in Y.
Y.shape
```

```python
# Returns the values of training and testing sets.
print("X Train Shape:", X_train.shape)
print("X Test Shape:", X_test.shape)
print("Y Train Shape:", Y_train.shape)
print("Y Test Shape:", Y_test.shape)
```

```python
# Now performing linear regression.
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, Y_train)
```

```python
# Predicting values of profit.
Y_predict = lr.predict(X_test)
Y_predict
```

```python
# Returns the length of the predicting values array.
len(Y_predict)
```

```python
# Actual profit values.
print(Y_test)
```

```python
# Returns the length of the actual values array.
len(Y_test)
```

```python
# Plotting a scatter plot between Y_predict and Y_test using matplotlib.
plt.scatter(Y_predict, Y_test)
```

```python
# Comparing actual values with predicted values.
comparison = pd.DataFrame({'Actual Values':Y_test, 'Predicted Values':Y_predict})
comparison
```

```python
# Pandas plotting of the graph between actual and predicted values.
comparison.plot(figsize = (5,5))
plt.show()
```

```python
# Plotting a bar graph with respect to acrual and predicted values.
comparison.plot(figsize=(10, 10), kind='bar')
plt.show()
```

```python
# Plotting a regplot using seaborn.
sns.regplot(x = 'Actual Values', y = 'Predicted Values', data = comparison, color = 'blue')
```

```python
# Importing r2_score.
from sklearn.metrics import r2_score
```

```python
# Calculating the accuracy.
score = r2_score(Y_test, Y_predict)
print(score)
```
